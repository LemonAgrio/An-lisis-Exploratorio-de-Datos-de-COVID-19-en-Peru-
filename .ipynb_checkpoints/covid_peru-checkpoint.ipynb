```python
# AnÃ¡lisis Exploratorio de Datos COVID-19 en PerÃº
# Autor: Roy AlarcÃ³n

import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
ruta = "C:/Users/WILSON/covid-peru-analysis/data/covid_peru.csv"
df = pd.read_csv(ruta, sep=';', encoding='utf-8')

# Eliminar columna vacÃ­a
if 'Unnamed: 8' in df.columns:
    df.drop(columns=['Unnamed: 8'], inplace=True)

# Limpieza de columna FECHA
df['FECHA'] = pd.to_datetime(df['FECHA'], dayfirst=True, errors='coerce')

# Limpieza de columna EDAD
df['EDAD'] = pd.to_numeric(df['EDAD'], errors='coerce')
df = df[df['EDAD'].between(0, 115)]

# -------------------------------
# VisualizaciÃ³n 1: DistribuciÃ³n de Edad
# -------------------------------
plt.figure(figsize=(10, 5))
df['EDAD'].hist(bins=50, color='skyblue', edgecolor='black')
plt.title('DistribuciÃ³n de Edad')
plt.xlabel('Edad')
plt.ylabel('Cantidad de casos')
plt.grid(True)
plt.tight_layout()
plt.show()

# -------------------------------
# VisualizaciÃ³n 2: Casos por AÃ±o
# -------------------------------
df['AÃO'] = df['FECHA'].dt.year
plt.figure(figsize=(8, 4))
df['AÃO'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Casos por AÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('Cantidad de casos')
plt.tight_layout()
plt.show()

# -------------------------------
# VisualizaciÃ³n 3: Casos por Departamento
# -------------------------------
plt.figure(figsize=(10, 6))
df['DEPARTAMENTO DOMICILIO'].value_counts().head(10).sort_values().plot(kind='barh', color='green')
plt.title('Top 10 Departamentos con mÃ¡s casos')
plt.xlabel('Cantidad de casos')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Agrupar edades por rangos
df['grupo_edad'] = pd.cut(df['EDAD'], bins=[0, 20, 40, 60, 80, 100, 120],
                          labels=['0-19', '20-39', '40-59', '60-79', '80-99', '100+'])

# GrÃ¡fico de barras
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='grupo_edad', hue='SEXO')
plt.title('DistribuciÃ³n de casos por grupo etario y sexo')
plt.xlabel('Grupo de edad')
plt.ylabel('Cantidad de casos')
plt.show()
```

```python
casos_por_fecha = df['FECHA'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
casos_por_fecha.plot()
plt.title('EvoluciÃ³n de casos en el tiempo')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de casos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv("C:/Users/WILSON/covid-peru-analysis/data/covid_peru.csv", sep=";")

# Eliminar columna innecesaria
df = df.drop(columns=["Unnamed: 8"], errors="ignore")

# Convertir a fecha
df["FECHA"] = pd.to_datetime(df["FECHA"], dayfirst=True, errors="coerce")

# Corregir EDAD
df["EDAD"] = pd.to_numeric(df["EDAD"], errors="coerce")

# Limpiar texto de columnas categÃ³ricas
cat_cols = ["SEXO", "TIPO SEGURO", "ESTADO CIVIL", "PAIS DOMICILIO", "DEPARTAMENTO DOMICILIO"]
for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.upper()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["EDAD"].dropna(), bins=30, kde=True)
plt.title("DistribuciÃ³n de la Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df["SEXO"].value_counts().plot(kind="bar", title="Casos por sexo")
plt.ylabel("Cantidad de casos")
plt.show()
```

```python
top_departamentos = df["DEPARTAMENTO DOMICILIO"].value_counts().head(10)
top_departamentos.plot(kind="barh", title="Top 10 departamentos con mÃ¡s casos")
plt.xlabel("Cantidad de casos")
plt.gca().invert_yaxis()
plt.show()

```

```python
df_fecha = df.dropna(subset=["FECHA"])
conteo_por_dia = df_fecha["FECHA"].value_counts().sort_index()

plt.figure(figsize=(15, 5))
conteo_por_dia.plot()
plt.title("EvoluciÃ³n diaria de casos reportados")
plt.xlabel("Fecha")
plt.ylabel("NÃºmero de casos")
plt.show()
```

```python
# AnÃ¡lisis de Datos COVID-19 en PerÃº

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ConfiguraciÃ³n general de grÃ¡ficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)

# 1. Carga de datos

# Leer CSV original
ruta = "C:/Users/WILSON/covid-peru-analysis/data/covid_peru.csv"
df = pd.read_csv(ruta, sep=';', encoding='utf-8')

# 2. Limpieza de datos

# Eliminar columnas innecesarias
if "Unnamed: 8" in df.columns:
    df = df.drop(columns=["Unnamed: 8"])

# Convertir fechas
df["FECHA"] = pd.to_datetime(df["FECHA"], dayfirst=True, errors='coerce')

# Convertir edad a numÃ©rico
df["EDAD"] = pd.to_numeric(df["EDAD"], errors='coerce')

# Normalizar texto en columnas categÃ³ricas
cat_cols = ["SEXO", "TIPO SEGURO", "ESTADO CIVIL", "PAIS DOMICILIO", "DEPARTAMENTO DOMICILIO"]
for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.upper()

# 3. AnÃ¡lisis exploratorio

## 3.1 DistribuciÃ³n de edad
sns.histplot(df["EDAD"].dropna(), bins=30, kde=True)
plt.title("DistribuciÃ³n de la Edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

## 3.2 Casos por sexo
df["SEXO"].value_counts().plot(kind="bar", title="Casos por sexo")
plt.ylabel("Cantidad de casos")
plt.show()

## 3.3 Casos por departamento
top_deps = df["DEPARTAMENTO DOMICILIO"].value_counts().head(10)
top_deps.plot(kind="barh", title="Top 10 departamentos con mÃ¡s casos")
plt.xlabel("Cantidad de casos")
plt.gca().invert_yaxis()
plt.show()

## 3.4 Casos por fecha
conteo_por_fecha = df.dropna(subset=["FECHA"])["FECHA"].value_counts().sort_index()
conteo_por_fecha.plot()
plt.title("EvoluciÃ³n diaria de casos reportados")
plt.xlabel("Fecha")
plt.ylabel("NÃºmero de casos")
plt.show()

# 4. Exportar CSV limpio
df.to_csv("C:/Users/WILSON/covid-peru-analysis/data/covid_peru_limpio.csv", index=False)

# 5. Conclusiones
# - La mayorÃ­a de casos se concentran en Lima.
# - La edad promedio estÃ¡ en adultos mayores (media â 67 aÃ±os).
# - Predominan los casos del sexo masculino.
# - Se observa un patrÃ³n claro de crecimiento y descenso en ciertas fechas clave.
```
